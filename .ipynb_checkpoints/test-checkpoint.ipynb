```python
import torch
import numpy as np
n = dist_matrix.shape[0]
#k = k_numb                         #how many centers
dist = torch.Tensor(n)  #각 class와 center class 간의 거리 
index_map = dict()                 #집합 요소 id와 그 요소의 center id  
initial_center_id = int(np.random.randint(n))  #초기 center class의 id 사용자가 커스터마이징 가능
centers = torch.Tensor([initial_center_id])
threshold = 25.
k=None
#centers = torch.Tensor([initial_center_id])        #center id list 초기화

#threshold =threshold               #클러스터의 최대 반지름
```

```python
#dist의 값을 모두 무한에 가깝게 만든다. 즉 첫 center의 탐색 범위를 무한으로 만든다.
for i in range(n): 
    dist[i] = torch.Tensor([10**9])
    index_map.update({i: initial_center_id})
```

```python
i=0
maxid = initial_center_id
while True:
    for j in range(n):                             #center와 점들 간의 거리를 업데이트
        if dist_matrix[maxid][j] < dist[j]:
            index_map[j] = maxid
            dist[j] = dist_matrix[maxid][j]
    maxid = int(torch.argmax(dist))
    centers = torch.cat((centers, torch.Tensor([maxid])), dim=0)
    i+=1
    if threshold != None  :                             #거리의 최대값을 지정한 경우
        if torch.max(dist) <= threshold:
            break
    if i== k:                                      #center 수의 값을 지정한 경우
        break    
```
